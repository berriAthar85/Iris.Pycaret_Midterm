```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score

# Load your dataset into a pandas dataframe
data = pd.read_csv('iris.csv')

# Set up the environment for modeling
exp_clf = setup(data, target = 'Species')

# Split the data into training and testing sets
X = data.drop('Species', axis=1)
y = data['Species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Train Logistic Regression model 1
lr_model1 = LogisticRegression(random_state=1)
lr_model1.fit(X_train, y_train)

# Train Logistic Regression model 2
lr_model2 = LogisticRegression(random_state=2)
lr_model2.fit(X_train, y_train)

# Train Logistic Regression model 3
lr_model3 = LogisticRegression(random_state=3)
lr_model3.fit(X_train, y_train)

# Create the ensemble model
ensemble_model = VotingClassifier(
    estimators=[('lr1', lr_model1), ('lr2', lr_model2), ('lr3', lr_model3)],
    voting='soft'
)

# Fit the ensemble model to the training data
ensemble_model.fit(X_train, y_train)

# Evaluate the ensemble model on the testing data
y_pred = ensemble_model.predict(X_test)
score = accuracy_score(y_test, y_pred)
print(f'Ensemble model accuracy: {score:.3f}')

```
